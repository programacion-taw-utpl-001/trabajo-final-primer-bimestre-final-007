```python
mes = list((datos_csv["mes_div"]))

meses = list(set(mes))
enero = mes.count(meses[5])
febrero = mes.count(meses[1])
marzo = mes.count(meses[2])
abril = mes.count(meses[3])
mayo = mes.count(meses[4])
junio = mes.count(meses[2])
julio = mes.count(meses[0])
agosto = mes.count(meses[7])
septiembre = mes.count(meses[8])
octubre = mes.count(meses[9])
noviembre = mes.count(meses[10])
diciembre = mes.count(meses[11])
colores=["#F39B36","#58F66D","#F61F46","#4A56F7","#F8F422","#D759EE","#C77F40","#BCE144","#64E7E7","#D4EF4D","#A0A4A9","#51EAA7"]
lista=[enero,febrero,marzo,abril,mayo,junio,julio,agosto,septiembre,octubre,noviembre,diciembre]
p = figure(x_range= meses, plot_width=1000,plot_height=300, title="Divorcios A Nivel Nacional por meses",title_location="below", toolbar_location="right")
p.vbar(x=meses, top = lista,width=0.6,color=colores)
p.xgrid.grid_line_color= None
p.y_range.start=0
show(p)
```

```python
provincias=['Azuay','Guayas','Loja','Pichincha']
```

```python
years=['2014','2015','2016']
```

```python
valores={'provincias':provincias,
         '2014':[49,132,11,166],
         '2015':[326,1131,105,1237],
         '2016':[1541,3966,485,5120]
    
}
```

```python
source = ColumnDataSource(valores)

p = figure(x_range=provincias, y_range=(0, 6000), plot_height=350, title="Divorcios en las Provincias")

p.vbar(x=dodge('provincias', -0.25, range=p.x_range), top='2014', width=0.2, source=source,
       color="#E7D108", legend=value("2015"))

p.vbar(x=dodge('provincias',  0.0,  range=p.x_range), top='2015', width=0.2, source=source,
       color="#02BEBB", legend=value("2016"))

p.vbar(x=dodge('provincias',  0.25, range=p.x_range), top='2016', width=0.2, source=source,
       color="#B20407", legend=value("2017"))

p.x_range.range_padding = 0.1
p.xgrid.grid_line_color = None
p.legend.location = "top_left"
p.legend.orientation = "horizontal"

show(p)
```
